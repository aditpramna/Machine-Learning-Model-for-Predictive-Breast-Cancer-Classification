```python
# Mengambil data pipeline
import pickle
import numpy as np
import pandas as pd

#Untuk mengabaikan fitur peringatan
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load All Files

with open('all_process_aditpram.pkl', 'rb') as file_1:
  all_process_aditpram = pickle.load(file_1)
```

```python
np.random.seed(7)

# Membuat data dummy dengan 1000 baris
n_samples = 1000     

# Menentukan rentang untuk masing-masing fitur
feature_ranges = {
    'radius_mean': (7, 28),
    'texture_mean': (9, 40),
    'perimeter_mean': (44, 189),
    'area_mean': (144, 2501),
    'smoothness_mean': (0.05, 0.16),
    'compactness_mean': (0.019, 0.35),
    'concavity_mean': (0, 0.43),
    'concave points_mean': (0, 0.20),
    'symmetry_mean': (0.11, 0.3),
    'fractal_dimension_mean': (0.05, 0.1),
    'radius_se': (0.1, 2.9),
    'texture_se': (0.36, 4.9),
    'perimeter_se': (0.76, 22),
    'area_se': (6.8, 542.2),
    'smoothness_se': (0.0017, 0.03),
    'compactness_se': (0.002, 0.14),
    'concavity_se': (0, 0.4),
    'concave points_se': (0, 0.05),
    'symmetry_se': (0.007, 0.08),
    'fractal_dimension_se': (0.0009, 0.03),
    'radius_worst': (7.9, 36),
    'texture_worst': (12, 50),
    'perimeter_worst': (50, 251),
    'area_worst': (185, 4254),
    'smoothness_worst': (0.07, 0.2),
    'compactness_worst': (0.03, 1.06),
    'concavity_worst': (0, 1.25200),
    'concave points_worst': (0, 0.29),
    'symmetry_worst': (0.16, 0.7),
    'fractal_dimension_worst': (0.06, 0.21)
}

# Membuat data dummy secara acak
dummy_data = {}
for feature, (min_val, max_val) in feature_ranges.items():
    dummy_data[feature] = np.random.uniform(min_val, max_val, n_samples)
```

```python
data_inf = pd.DataFrame(dummy_data)
data_inf
```

```python
# Melihat isi pipeline
all_process_aditpram
```

```python
# Melakukan prediksi data dummy
y_pred_inf = all_process_aditpram.predict(data_inf)
y_pred_inf
```

```python
# Convert into DataFrame

y_pred_inf_df = pd.DataFrame(y_pred_inf, columns=['Price - Prediction'])
y_pred_inf_df.value_counts()
```

```python
# Merge Inference-Set and Prediction

tabel_inference = pd.concat([data_inf, y_pred_inf_df], axis=1)
tabel_inference
```
